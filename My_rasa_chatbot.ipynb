# 2.IMPORT THE NECESSARY LIBRARIES.
import os(to enable you to create the file directory)

```python
import os
import rasa
import nest_asyncio

nest_asyncio.apply()
print("event loop ready")
```

# 3. CREATE YOUR PROJECT DIRECTORY.


```python
from rasa.cli.scaffold import create_initial_project
project = "my_first_chatbot"
create_initial_project(project)
```

```python
os.chdir(project)
print(os.listdir("."))
#
```

# 4. ASSIGNING VARIABLES TO THE FILES CREATED.

```python
config = "config.yml"
training_files = "data/"
domain = "domain.yml"
output = "models/"
print(config, training_files, domain, output)
```

# 5.TRAIN YOUR RASA MODEL.

```python
model= rasa.train(domain, config, [training_files], output)
print(model)
```

# 6.TEST YOUR CHAT BOT


```python
from rasa.jupyter import chat

chat(model.model)
```
